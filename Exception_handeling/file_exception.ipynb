```python
file = None

try:
    file = open("example1.txt", "r")
    content = file.read()
    print(content)

except FileNotFoundError:
    print("The file 'example1.txt' was not found.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    if file is not None and not file.closed:
        file.close()
        print("file closed")
```
